```python
import time
t0 = time.clock()
import pandas as pd
from pandas.tseries.offsets import BDay
import numpy as np
import datetime as dt
from datetime import time as time
from copy import copy
import warnings
warnings.filterwarnings('ignore',category=pd.io.pytables.PerformanceWarning)
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy.stats import linregress
import seaborn as sns

```

```python
#SCRIPT TO SCRAPE INTRADAY FINANCIAL DATA:
# datetime management

d = dt.date.today()
# ---------- Days ---------- 
l10 = d - 10 * BDay()
l21 = d - 21 * BDay()
l63 = d - 63 * BDay()
l252 = d - 252 * BDay()
# ---------- Years ---------- 
l252_x2 = d - 252 * 2 * BDay() 
l252_x3 = d - 252 * 3 * BDay() 
l252_x5 = d - 252 * 5 * BDay()
l252_x7 = d - 252 * 7 * BDay() 
l252_x10 = d - 252 * 10 * BDay() 
l252_x20 = d - 252 * 20 * BDay() 
l252_x25 = d - 252 * 25 * BDay() 
# ================================================================== #

# filepath management

project_dir = 'holdings-spy.xls' 
price_path = project_dir + 'Stock_Price_Data\\'

#Make API Call
apikey = " "
def construct_barChart_url(sym, start_date, freq, api_key=apikey):
    '''Function to construct barchart api url'''
    
    url = 'http://marketdata.websol.barchart.com/getHistory.csv?' +\
            'key={}&symbol={}&type={}&startDate={}'.format(api_key, sym, freq, start_date)
    return url

def get_minute_data():
    '''Function to Retrieve <= 3 months of minute data for SP500 components'''
    
    # This is the required format for datetimes to access the API
    # You could make a function to translate datetime to this format
    start = '20180319000000'
    #end = d
    freq = 'minutes'    
    prices = {}
    symbol_count = len(syms)
    N = copy(symbol_count)
    try:
        for i, sym in enumerate(syms, start=1):
            api_url = construct_barChart_url(sym, start, freq, api_key=apikey)
            try:
                csvfile = pd.read_csv(api_url, parse_dates=['timestamp'])
                csvfile.set_index('timestamp', inplace=True)
                prices[sym] = csvfile
            except:
                continue
            N -= 1
            pct_total_left = (N/symbol_count)
            print('{}..[done] | {} of {} symbols collected | percent remaining: {:>.2%}'.format(\
                                                                sym, i, symbol_count, pct_total_left)) 
    except Exception as e: 
        print(e)
    finally:
        pass
    px = pd.Panel.from_dict(prices)

    return px

# header=3 to skip unnecesary file metadata included by State Street    
spy_components = pd.read_excel(project_dir, header=3)
syms = spy_components.Identifier.dropna()
syms = syms.drop(syms.index[-1]).sort_values()

pxx = get_minute_data()
```

```python
# Begin Plotting and Conducting Regression Analysis
Y = fin_data_FB['open']
X = fin_data_FB['Compound_Score']

#Regression Analysis
r_square = linregress(X.astype('float64'),Y.astype('float64'))
print(r_square)

# plot points 
sns.set(color_codes=True)
sns.regplot(X,Y,scatter = True)

plt.xlabel("Mean Compound Score per minute during market hours")
plt.ylabel("Price of Facebook Stock(OPEN) in $")
plt.title("Correlational Analysis between Twitter Sentiment and Facebook Stock on 3/19/18")
plt.show()
```
