```python
# Dependencies
import csv
import tweepy
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#Read in twitter data from CSV
x = pd.read_csv('output.csv',error_bad_lines=False) 
#x.values
```

```python
#Clean up data to acquire the tweets, the date and time of the tweets:
tweets = []
times = []
dates = []

#Make array a list
y = x.values.tolist()

#Split string in list to get the tweet and the time
for i in y:
    for j in i:
        v = j.split(";")
        b = v[1].split(" ")
        dates.append(b[0])
        times.append(b[1])
        tweets.append(v[4])
        


    
```

```python
#Tweet analysis made into a dictionary:
sentiment = []
counter = 0
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

for tweet in tweets:
        # Run Vader Analysis on each tweet
        compound = analyzer.polarity_scores(tweet)["compound"]
        pos = analyzer.polarity_scores(tweet)["pos"]
        neu = analyzer.polarity_scores(tweet)["neu"]
        neg = analyzer.polarity_scores(tweet)["neg"]

        # Add each value to the appropriate array
        compound_list.append(compound)
        positive_list.append(pos)
        negative_list.append(neg)
        neutral_list.append(neu)
        sentiment.append({"Compound": np.mean(compound_list),
                          "Positive": np.mean(positive_list),
                          "Neutral": np.mean(negative_list),
                          "Negative": np.mean(neutral_list),
                          })
        
        
```

```python
#Create Dataframe with sentiment data
sentiment_pd = pd.DataFrame(sentiment)
sentiment_pd["Times"] = times
sentiment_pd["Dates"] = dates
sentiment_pd.head()
```

```python
#Create sub dataset with relevant data for market hours
sentiment_pd =  sentiment_pd.sort_values(by = ["Times"])
sentiment_pd = sentiment_pd.reset_index()
sentiment_pd.drop(['index'],axis = 1)
dataset = sentiment_pd.iloc[51723:114688,:]
dataset.to_csv("Tweet_DF")
```
